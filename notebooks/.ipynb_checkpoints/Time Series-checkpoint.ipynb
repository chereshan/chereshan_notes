```python
ts_length = 100
ϵ_values = []   # empty list

for i in range(ts_length):
    e = np.random.randn()
    ϵ_values.append(e)

plt.plot(ϵ_values)
plt.show()
```

```python
ts_length = 100
ϵ_values = []
i = 0
while i < ts_length:
    e = np.random.randn()
    ϵ_values.append(e)
    i = i + 1
plt.plot(ϵ_values)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
def generate_data(n):
    ϵ_values = []
    for i in range(n):
        e = np.random.randn()
        ϵ_values.append(e)
    return ϵ_values

data = generate_data(100)
plt.plot(data)
plt.show()
```

```python
def generate_data(n, generator_type):
    ϵ_values = []
    for i in range(n):
        if generator_type == 'U':
            e = np.random.uniform(0, 1)
        else:
            e = np.random.randn()
        ϵ_values.append(e)
    return ϵ_values

data = generate_data(100, 'U')
plt.plot(data)
plt.show()
```

### Пример: Баланс банка во времени

* Нет снятий в течение периода времени, последняя дата обознчается как $T$.
* $b_0$ - изначальный баланс
* $r$ - ставка процента
* Баланс обновляется в период с $t$ к $t+1$ следующим образом: $b_{t+1}=(1+r)b_t$

```python
r = 0.025         # interest rate
T = 50            # end date
b = np.empty(T+1) # an empty NumPy array, to store all b_t
b[0] = 10         # initial balance

for t in range(T):
    b[t+1] = (1 + r) * b[t]

plt.plot(b, label='bank balance')
plt.legend()
plt.show()
```

#### Задача 1.  Генерация процесса Маркова AR(1)

```python
import numpy as np
import matplotlib.pyplot as plt
T=200
alpha=0.9
x=[0]
for i in range(1,T):
    x.append(x[i-1]*0.9+np.random.randn())
    
plt.plot(np.linspace(0,200,200),x)
```

#### Задача 2. 
AR(1): **$x_t=\varphi x_{t-1} + \epsilon_{t}$** c различными $\varphi$

```python
def ts_sim(alpha,T=200):
    x=[0]
    for i in range(1,T):
        x.append(x[i-1]*alpha+np.random.randn())
    return x
plt.plot(ts_sim(0))
```

```python
plt.plot(ts_sim(0.98,T=200))
```

#### Задача 3. AR(1) c модулем

```python
def ts_sim_abs(alpha,T=200):
    x=[0]
    for i in range(1,T):
        x.append(abs(x[i-1])*alpha+np.random.randn())
    return x
plt.plot(ts_sim_abs(0.9))
```
