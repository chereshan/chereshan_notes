## Введение 

В Python есть довольно много GUI фреймворков (graphical user interface), однако только Tkinter встроен в стандартную библиотеку языка. У Tkinter есть несколько преимуществ. Он кроссплатформенный, поэтому один и тот же код можно использовать на Windows, macOS и Linux. 

Хотя Tkinter является популярным GUI фреймворком на Python, у него есть свои недостатки. Один из них заключается в том, что графические интерфейсы, созданные с использованием Tkinter, выглядят устаревшими. Если вам нужен современный, броский интерфейс, то Tkinter может оказаться не совсем тем, для этого есть PyQt5 который развивается сильнее в данном плане.

Tk - это библиотека виджетов, имеющая корни в языке программирования Tool Command Language (Tcl). Оба эти инструмента были разработы John Ousterhout в 1980-ых для упрощения программирования для инженеров. Как и сам Python Tcl/Tk возник на Unix-системе и уже позже мигрировал на другие ОС такие, как Windows и macOS. 

**Tkinter** - это Python интерфйс *Tk* GUI библиотеки. 

**Преимущества Tkinter:**
* Это стандартная библиотека 
* Он стабилен - у него медленный девеломпмент и эволюционный
* Это всего лишь GUI toolkit - он не имеет threading library, network stack или filesystem API
* Он простой и понятный. 

**Негативные аспекты Tkinter:**
* Он выглядит и ощущается устаревшим.
* Не способен на более комплексные виджеты. Не имеет:rich text editors, 3D graphics embedding, HTML viewers, или specialized input widgets

````python
"""Hello World application for Tkinter""" #docstring - каждый скрипт Python должен начинаться с этой строки. Это название и описание программы.
import tkinter as tk
root=tk.Tk()
# root window - корневое окно, которое должно иметь каждое приложени tkinter
label=tk.Label(root, text='Hello world') # Виджет с названием. 
# Первый аргумент - это ВСЕГДА parent widget (master widget) (родительский виджет)
# родительский виджет - эт овиджет, на который расположится Label 
# Виджеты в tkinter ранжированы иерархично, каждый виджет содержиться в другом вплоть до корневого окна
# мы также передали корневое слово - text
# Когда мы создали виджет его надо распаковать в GUI 
label.pack() 
# Метод .pack() - это **geometry manager method**. Его задачи в том, чтобы определит то, как виджет будет присоединен к родителському виджету и нарисовать его. Без него виджет будет существовать, но его не будет видно на окне
root.mainloop() # последняя строка программы. она начинает event loop приложения.
# **event loop** - это бесконечный цикл, который непрерыыно процессит любой ивент, который происходит во время execution этой программы.
# Ивенты: движения мышкой, нажатия или другие сгенерированные юзером активности. 
# Этот цикл продолжается до тех пор, пока программа не выйдет, так что любой код после этой строки не будет запуще, пока главное окно программы не закрыто. По этой причине данная строка обычно последняя
````

* ``root=tk.Tk()`` - создает **root window**, корневое окно, которое должно иметь каждое приложение tkinter


* ``label=tk.Label(root, text='Hello world')`` - виджет с названием. Первый аргумент (здесь это ``root``) - это всегда **parent widget** (*master widget*), т.е. *родительский виджет*


* **Родительский виджет** - это виджет, на котором ... Виджеты в tkinter ранжированы иерархично, каждый виджет содержиться в другом вплоть до корневого окна


* Мы также передали в функцию ``Label`` ключевое слово - text


* ``label.pack()`` - метод ``.pack()`` распоковывает элемент в GUI. Данный метод также является т.н. **geometry manager method** Его задача в том, чтобы определить то, как виджет будет присоединен к родителському виджету и нарисовать его. Без него виджет будет существовать, но его не будет видно на окне


* ``root.mainloop()`` - как правило последняя строка практически любой приложения tkinter. Она начинает **event loop** приложения.
* **event loop** - это бесконечный цикл, который непрерывно процессит любой *ivent* (событие), который происходит во время execution этого приложения.
* События: движения мышкой, нажатия или другие сгенерированные пользователем виды активности. 
* Этот цикл продолжается до тех пор, пока программа не выйдет, так что любой код после этой строки не будет запущен, пока главное окно программы не закрыто. По этой причине данная строка обычно последняя.

````python
# banana_survey.py
"""A banana preferences survey written in Python with Tkinter"""
import tkinter as tk
root = tk.Tk()
title = tk.Label(root,  
                 text='Please take the survey',
                 font=('Arial 16 bold'),
                 bg='brown',fg='#FF0')
# set the title
name_label = tk.Label(root, text='What is your name?')
name_inp = tk.Entry(root)

eater_inp = tk.Checkbutton(
 root,
 text='Check this box if you eat bananas'
)
num_label = tk.Label(
 root,
 text='How many bananas do you eat per day?'
)
num_inp = tk.Spinbox(root, 
                     from_=0,
                     to=1000, increment=1)
#====================================================
color_label = tk.Label(
 root,
 text='What is the best color for a banana?'
)
color_inp = tk.Listbox(root, height=2) # Only show selected item
# add choices
color_choices = (
 'Any', 'Green', 'Green-Yellow',
 'Yellow', 'Brown Spotted', 'Black'
 )
for choice in color_choices:
    color_inp.insert(tk.END, choice)
#====================================================
root.title('Banana interest survey')
# set the root windowsize
root.geometry('640x480+300+300')
root.resizable(False, False)

title.pack()
name_label.pack()
name_inp.pack()
eater_inp.pack()
num_label.pack()
num_inp.pack()
color_label.pack()
color_inp.pack()



root.mainloop()
````

````python
# banana_survey.py
"""A banana preferences survey written in Python with Tkinter"""
import tkinter as tk
root = tk.Tk()
title = tk.Label(root,  
                 text='Please take the survey',
                 font=('Arial 16 bold'),
                 bg='brown',fg='#FF0')
# set the title
name_label = tk.Label(root, text='What is your name?')
name_inp = tk.Entry(root)

eater_inp = tk.Checkbutton(
 root,
 text='Check this box if you eat bananas'
)
num_label = tk.Label(
 root,
 text='How many bananas do you eat per day?'
)
num_inp = tk.Spinbox(root, 
                     from_=0,
                     to=1000, increment=1)

color_label = tk.Label(
 root,
 text='What is the best color for a banana?'
)
color_inp = tk.Listbox(root, height=2) # Only show selected item
# add choices
color_choices = (
 'Any', 'Green', 'Green-Yellow',
 'Yellow', 'Brown Spotted', 'Black'
 )
for choice in color_choices:
    color_inp.insert(tk.END, choice)
    
#====================================================    
    
plantain_label = tk.Label(root,
                         text='Do you eat plantains?')
plantain_frame = tk.Frame(root)
plantain_yes_inp = tk.Radiobutton(plantain_frame,
                                  text='Yes')
plantain_no_inp = tk.Radiobutton(plantain_frame,
                                 text='Ewww, no!')
#====================================================
root.title('Banana interest survey')
# set the root windowsize
root.geometry('640x480+300+300')
root.resizable(False, False)

title.pack()
name_label.pack()
name_inp.pack()
eater_inp.pack()
num_label.pack()
num_inp.pack()
color_label.pack()
color_inp.pack()

plantain_label.pack()
plantain_frame.pack()
plantain_yes_inp.pack()
plantain_no_inp.pack()


root.mainloop()
````

````python
# banana_survey.py
"""A banana preferences survey written in Python with Tkinter"""
import tkinter as tk
root = tk.Tk()
title = tk.Label(root,  
                 text='Please take the survey',
                 font=('Arial 16 bold'),
                 bg='brown',fg='#FF0')
name_label = tk.Label(root, text='What is your name?')
name_inp = tk.Entry(root)
eater_inp = tk.Checkbutton(root,text='Check this box if you eat bananas')
num_label = tk.Label(root,text='How many bananas do you eat per day?')
num_inp = tk.Spinbox(root,from_=0,to=1000, increment=1)

color_label = tk.Label(root,text='What is the best color for a banana?')
color_inp = tk.Listbox(root, height=2) # Only show selected item
# add choices
color_choices = (
 'Any', 'Green', 'Green-Yellow',
 'Yellow', 'Brown Spotted', 'Black'
 )
for choice in color_choices:
    color_inp.insert(tk.END, choice)
    
plantain_label = tk.Label(root,text='Do you eat plantains?')
plantain_frame = tk.Frame(root)
plantain_yes_inp = tk.Radiobutton(plantain_frame,text='Yes')
plantain_no_inp = tk.Radiobutton(plantain_frame,text='Ewww, no!')
#====================================================    
banana_haiku_label = tk.Label(
 root,
 text='Write a haiku about bananas')
banana_haiku_inp = tk.Text(root, height=3)
submit_btn = tk.Button(root, text='Submit Survey')
output_line = tk.Label(root, text='', anchor='w', justify='left')
#====================================================
root.title('Banana interest survey')
# set the root windowsize
root.geometry('640x480+300+300')
root.resizable(False, False)

title.pack()
name_label.pack()
name_inp.pack()
eater_inp.pack()
num_label.pack()
num_inp.pack()
color_label.pack()
color_inp.pack()
plantain_label.pack()
plantain_frame.pack()
plantain_yes_inp.pack()
plantain_no_inp.pack()
banana_haiku_label.pack()
banana_haiku_inp.pack()
submit_btn.pack()
output_line.pack()

root.mainloop()
````
