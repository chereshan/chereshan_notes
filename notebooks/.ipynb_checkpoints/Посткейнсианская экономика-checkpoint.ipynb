**Introduction**

> **Введение**

The publication of John Maynard Keynes’ General Theory of Employment, Money and
Interest in 1936 marked a turning point in the history of economic thought.

> Публикация Джоном Мейнардом Кейнсом его *Общей теории занятости, процента и денег* в 1936 году ознаменовала поворотный момент в истории экономической мысли.

It was written in response to the Great Depression and provided a theoretical framework for
understanding financial crises and involuntary unemployment.

> Эта работа была написана как реакция на Великую Депрессию и предоставила теоретический фреймворк для понимания финансовых кризисов и вынужденной безработицы. 

Politically, this lent itself
to a justification for government intervention and financial regulation: if capitalist market
economies are unstable and prone to generate periods of high unemployment, governments should stabilise the economy by running budget deficits in times of crises
and regulate finance to prevent the emergence of bubbles.

> Политически, это послужило обоснованием для государтсвенных интервенций и финансовых регуляций: если капиталистическая рыночная экономика нестабильна и склонна к порождению периодов высокой безработицы, то правительствам следует стабилизировать экономику путем бюджетных дефицитов во времена кризисов и регулирования финансового сектора для предотвращения возникновения пузырей.``

This provided a reference
point for the socially inclusive economic policy in the postwar era. 

> Это стало точкой отсчета для инклюзивной экономической политики послевоенной эры. 

During the 1980s,
Keynesianism was abandoned and policy makers around the world switched to ‘unleash
capitalism’, which resulted in the global financial crisis in 2008. Since then, Keynesian
theory has met renewed interest.

> В течение 1980-х кейнсианство было заброшено и политические деятели по всему миру переключились на "развязывание рук капитализму", что в результате привело к глобальному финансовому кризису 2008 года. С тех пор кейнсиансткая теория привлекает возобновленный интерес. 

In terms of economic theory, the Keynesian revolution has led to several very different responses. 

> В терминах экономической теории, Кейнсианская революция привела к ряду различных ответных реакций.

First, modern mainstream (‘neoclassical’) economics grew out of the
rejection of Keynes and has resurrected orthodox ‘supply-side’ economics, which is generally hostile to government intervention. 

> Во-первых, современная мейнстримная (*неоклассическая*) экономика выросла из отрицания Кейнса и воскресила ортодоксальную экономику "стороны предложения", которая в целом воинственно настроенна к государсвтенным интервенциям.

Second, some economists have tried to marry
Keynes and neoclassical economics, which is often called the Neoclassical-Keynesian
Synthesis. 

> Во-вторых, некоторые экономисты пытались "пожинить" Кейнса и неоклассическую экономику, что получило название Неоклассического-Кейнсианского синтеза. 

Third, it led to the development of Post-Keynesian Economics (PKE), a
school of thought that highlights the distinctiveness of Keynes’ approach. 

> В-третьих, оно привело к разработке Посткейнсиансклй экономики (Post-Keynesian Economics, PKE), школе мысли, которая подчеркивает самобытность кейнсианского подхода. 


PKE takes
Keynes’ break from orthodox economics as its starting point, but it has on several issues
gone beyond Keynes’ own analysis.

> PKE берет разрыв Кейнса с ортодокальной экономикой в качестве своей исходной точки, но во многих вопросах заходит дальше кейнсианского анализа. 

The Keynesian revolution, in the PKE view, did
not go far enough. 

> Кейнсианская революция с точки зрения PKE не зашла достаточно далеко. 

In particular, PKE has incorporated the analysis of class relations
and developed further the role of income distribution and the theory of endogenous
money and financial instability, as discussed below. 

> В частности, PKE инкорпорировала в свой анализ классовые отношения и развила далее роль распределения дохода, теорию эндогенных денег и финансовой нестабильности, о которых будет говориться ниже. 

Next to Keynes, Michał Kalecki,
Joan Robinson, Nicholas Kaldor and Hyman Minsky have been important proponents
within this vein.

> Следующими за Кейнсем, Michał Kalecki, Joan Robinson, Nicholas Kaldor и Hyman Minsky были важными сторонниками в данной стези. 


The Post-Keynesian view of how capitalism works is in sharp contrast to the neoclassical view: capitalist market economies have weak self-stabilising mechanisms; they lead to
involuntary unemployment and exhibit frequent crises. 

> Посткейнсианский взгляд на то, как работает капитализм в резком контрасте с неоклассическим: капиталистические рыночные экономики имеют слабые самостабилизирующие механизмы; они ведут к вынужденной безработице и имеют тенденцию к частым кризисам. 

Theoretically, the most important feature of PKE for macroeconomic analysis is the principle of effective demand, which
states that expenditures (or demand) will determine output (e.g. economic growth), as
savings adjusts via the ‘multiplier’ process. 

> Теоретически, самая важная особенность PKE в рамках макроэкономического анализа - принцип эффективного спроса, который утверждает, что расходы (или спрос) определяет выпуск (или экономический рост), в то время как сбережения постраиваются через процесс мультипликатора.  

PKE has developed a distinct growth theory and distinguishes between wage-led and profit-led demand regimes. 

> PKE разработала отдельную теорию экономического роста, в рамках которой различаются два режима спроса: следующий за зарплатой режим (wage-led) и следующий за прибылью режим (profit-led).


It argues that
(a) financial markets are prone to instability and will, if left on their own, lead to boombust cycles; (b) money is created by banks as a side effect of their lending decisions;
(c) involuntary unemployment is a normal feature of labour markets; and (d) wage cuts
and structural reforms cannot cure unemployment. 

> Она утверждает, что 
(а) финансовые рынки склонны к нестабильности и оставленные самим себе ведут к Boom-Bust циклам;
(б) деньги создаются банковской системой как побочный эффект своих решений по кредитованию
(в) вынужденная безработица - нормальная черта рынков труда
(г) снижение зарплаты и структурные реформы не решают проблемы безработицы


This chapter will explain these building blocks and core theoretical arguments of PKE, after briefly touching upon its historical development.

> Данная глава объяснит эти строительные блоки и ядро теоретических аргументов PKE недолго после кратко коснемся его исторического развития. 

**Streams of Post-Keynesian Economics
and differences to mainstream Keynesians**

> **Направления посткейнсианской экономики ее отличия от мейнстримного кейнсианства**

The term Post-Keynesian Economics was coined in the 1970s; however, the intellectual
history of PKE is older than that. 

> Термин посткейнсианская экономика введен в 1970-ых; однако интеллектуальная история PKE длиннее этого. 

The first generation of Post-Keynesians – Joan Robinson, Nicholas Kaldor, Richard Kahn and Michał Kalecki – were mostly part of the Cambridge Circle, a group of economists who worked closely with Keynes at the University
of Cambridge. 

> Первое поколение посткейнсианцев - Joan Robinson, Nicholas Kaldor, Richard Kahn and Michał Kalecki - были в основном частью Кембриджского круга, группы экономистов, которые близко работали с Кейнсом в Кембрижском университете. 

To properly understand the development of PKE we first have to understand the changing reception of Keynesian arguments within the dominant economic
schools of the day. 

> Чтобы должным образом понять развитие PKE для мы должны понять измение восприятия кейнсианских аргументов в рамках доминирующих экономических школ своего времени. 

Soon after the publication of the General Theory, there were attempts
to marry neoclassical economics and Keynesian theory, which came to be known as the
Neoclassical-Keynesian Synthesis. 

> Вскоре после публикации Общей теории Кейнса существовали попытки "пожинить" неоклассическую экономику и кейнсианскую теорию, которые стали изветсны как неоклассический-кейнсианский синтез. 


It was led by John Hicks (1937), who developed the
‘IS-LM model’, which was popularised by Paul Samuelson’s widely used textbook and
remains a mainstay of economics textbooks to this day. 

> Во главе этого движения был Джон Хикс (1937), который разработал модель IS-LM, которая была популяризована широко используемым учебником Пола Самуэльсона и остается одной из твердынь в экономических цчебниках по сей день. 

Ultimately the Synthesis tried to
re-interpret Keynes as a special case of general equilibrium theory with wage and price
‘rigidities’. 

> В конечном счете синтез пытлася реинтерпретировать Кейнса как особый случай теории общего экономического равновесия с жетскостями (ригидности) цен и зарплат. 

The Synthesis approach created the division of economics into Macroeconomics, where Keynesian ideas played a prominent role, and Microeconomics, which
was neoclassical terrain.

> Подход синтеза создал раскол экономической теории на макроэкономику, где кейнсианские идеи играли заметную роль, и микроэкономику, которая оставалась неоклассической. 

Within Macroeconomics the short-run/long-run distinction plays a key role in reconciling the Keynesian and classical elements. 

> Внутри макроэкономики различие между краткосрочным и долгосрочным периодами играло ключевую роль в согласовании кейнсианских и классических элементов.

In the short run, the world works largely
along Keynesian lines: demand determines output and involuntary unemployment is
caused by lack of ‘effective demand’. 

> В краткосрочном периоде мир работает переимущественно по кейнсианских лекалам: спрос определяет выпуск и вынужденную безработицу, вызываемую энедостатком эффективного спроса.

This is due to ‘rigidities’ like sticky prices and
inflexible wages.

> Это связано с жесткостями вроде жестких цен и негибкими зарплатами. 

Prices at restaurants are a common example – supply and demand might
change constantly, but printing new menus is expensive so prices can stay ‘stuck’ for
some time.

> Цены в ресторанах - это распространенный пример. Предложение и спрос изменяются постоянно, но печать новых меню обходится недешево и потому цены могут "застривать" на некоторые периоды времени. 

However, in the long run, when prices and wages have time to adjust, the
economy should function according to neoclassical theory. 

> Однако в долгосрочном периоде, когда цены и зарплаты имеют время для подстройки, экономика будет функционировать согласно неоклассической теории. 

In the Synthesis, Keynes was
reduced to a special case, if a practically important one, of general equilibrium theory.

В синтезе, Кейнс был сведен к специальному случаю, хотя и важного, общего экономического равновесия.

In this period Post-Keynesians were mostly centred in Cambridge and criticised the
Synthesis, or as Joan Robinson called them the ‘bastard-Keynesians’, for reducing Keynes
to wage and price rigidities.

> В этот период посткейнсианцы были в основном сконцентрированы в Кембридже и критиковали синтез, или как Джоан Робинсон его называла, "бастардизированное кейнсианство" за его сведения Кейнса к жесткостям цен и зарплат. 

Keynes did argue that wages were in fact inflexible, but as
explained later, wage inflexibility is not the reason for unemployment

> Кейнс действительно утверждао, что зарлаты ынегибкие, но как позже будет объяснено, негибкость зарплат не является причиной безработицы.

In the 1950s and 60s PKE tried to generalise the short-term Keynesian analysis and
develop an independent analysis of the long period and the distribution of income. 

> В 1950-ых и 60-ых PKE пыталось обобщить краткосрочный кейнсианский анализ и развить независимый анализ долгосрочного периода и распределения дохода. 

The first-generation PK growth models have two distinct features. 

> Первое поколение посткейнсианских теория экономического роста имеют две отчетливые черты.

First, ‘animal spirits’ –
capitalists’ willingness to invest – play a key role as investment is an independent variable
in the long run (as opposed to being determined by the rate of savings, as in neoclassical
theory).

> Во-первых, "животные духи" - капиталистическое желание инвестировать - играют ключевую роль, т.к. инвестиции - независимая переменная в долгосрочном периоде (в отличие от ее зависимости от нормы сбережения в неоклассической теории)

Second, the distribution of income becomes a key adjusting variable, because
wage and profit income are associated with different saving and spending rates, or ‘marginal propensities to consume’. 

> Во-вторых, распределение дохода становится ключивуой подстраивающей переменной, потому что зарплаты и прибыли связаны с разными нормами сбережений и расходов, или предельной склонностью к потреблению. 

When someone spends money, it gets passed around the
economy many times after the initial transaction. 

> Когда кто-то тратит деньги, они циркулируют в экономике множество раз после изначальной транзакции. 

In this sense, new spending has both
an initial effect on effective demand (from the original purchase) and a ‘multiplier effect’
from all the additional purchases made possible by passing around the original money.

> В этом смысле новые расходы оказывают как изначальный эффект на эффективнй спрос (от изначальной покупки), так и мультипликативный эффект от всех последующих покупок, которые стали возможны циркуляцией денег от изначальной покупки. 

If workers spend a bigger portion of their income than capitalists, then wages should
also have a higher ‘multiplier’ than profits, and increasing wages (even at the expense of
profits) could be a way to increase effective demand.

> Если работники тратят большую долю своих доходов, чем капиталисты, то зарплаты будут иметь больший мультипликативный эффект, чем прибыли. И повышение зарплат (даже в ущерб прибылям) могут быть способом повышения эффективного спроса. 

In this period PKEs also criticised
the neoclassical ‘marginal product theory’ of income distribution as logically incoherent
in the so-called Cambridge Capital Controversies

> В данный период PKE такде критиковали неоклассическую теорию предельного дохода в распределении как логически непоследовательную в т.н. Кэмбриджском споре о капитале. 

In the 1970s mainstream economics changed, and by 1980 the so-called bastardKeynesians found themselves outside of the new mainstream, as the new schools of
‘Monetarism’ and ‘New Classical Economics’ challenged even the watered-down parts
of Keynesianism left in the Synthesis of the ’50s and ’60s. 

> В 1970-ых мейнстримная экономика изменилась, и к 1980-м т.н. "бастрардизированные кейнсианцы" обнаружили себя за пределами нового мейнстрима, т.к. новые школы Монетаризма и Новой классической экономики бросили вызов даже разбавленным кейнсианским частями оставшимся в синтезе 50ых и 60ых. 

This was part of the rise of
neoliberalism and resulted in a narrowing of the methods and theories acceptable in the
economics profession. 

> Данное явление было частью подъема неолиберализма и привела в результате к сужению множества допустимых методов  теория в экономической профессии.

Macroeconomics now had to be derived from strict microfoundations based on optimising behaviour, and markets were assumed to clear. 

> Макроэкономика теперь должны была быть получена из строгих микроэкооснований, исходящих из оптимизирующего поведения, а рынки предполагались расчищенными. 

For critical
economists life became more difficult.

> Для критически настроенных экономистов жизнь стала сложнее. 

That is why PKE (as well as other heterodox economics streams) began to institutionalise in the 1970s and ’80s by setting up their own scholarly associations and academic
journals – the established journals wouldn’t publish their work anymore. 

> Это то почему PKE (как и другие неортодокасльные направления экономики) начала институционализирвоаться в 1970-ых и 80-ых заводя свою собственную научную ассоциацию и академические журналы (уже сущесвтующие журналы не публиковали их работы).

This is the
time when PK academic journals like the Cambridge Journal of Economics and the Journal
of Post Keynesian Economics were founded.

> В это время были основаны такие академические журналы как Cambridge Journal of Economics и Journal of Post Keynesian Economics.

At the same time there was a geographical
shift. 

> В тоже время произошел географический сдвиг.

When the first generation of PKs retired at Cambridge, they were replaced by
mainstream economists. 

> Когда первое поколение посткейнсианцев уходила на пенсию в Кембридже, им на смену приходили мейнстримные экономисты. 

PKE lost its centre, but it gained breath; there were now PK
traditions and networks in several countries, and the USA started to play a more prominent role

> PKE потеряла свой прежний центр притяжения, но получила новое дыхание; возникли посткейнсианские традиции и сети в нескольких странах таких, как США, где они начали играть бОльшую роль. 

With the political changes and the New Classical counterrevolution there was a new
generation of PKs working on different topics. 

> С политическими изменениями и Новой Классической Контрреволюцией существовало новое поколение посткейнсианцев, работающих по различным экономическим проблемам. 

In response to Monetarism, the theory
of ‘endogenous money’ was developed; the concept of ‘fundamental uncertainty’ was
elaborated upon. 

> В ответ монетаризму, была разработана теория эндогенных денег; концепция фундаментальной неопределенности была доработана.

In the 1980s there was also a shift to more short- to medium-term
models; PKE developed the ‘wage-led growth’ models and there was more applied and
policy-oriented work. 

> В 1980-ых шел также сдвиг к более краткосрочным и среднесрочным моделям; PKE разработали модели wage-led экономического роста, также была разработаны приложения посткейнсианства к экономической политике. 

Financial issues became more prominent.

> Финансовая проблематика стала более заметной. 

The 1980s and ’90s also saw a new version of mainstream Keynesians. 

> 1980-е и 90-ые также засвидетельствовали новую версию мейнстримного кейнсианства. 

The so-called
New Keynesians accepted methodological individualism and the need for ‘microfoundations’ of macroeconomic models, but they argued that optimising behaviour does not
necessarily imply market clearing and full employment. 

> Т.н. Новые Кейнсианцы принимают методологический индивиуализм и необходимость в микрооснованияз макроэкономических моделей, но утверждают, что оптимизационное поведение необходимо не расчищает рынка и не обеспечивает полной занятости. 

Thus were developed theories
of ‘rational wages’ and ‘price stickiness’, based on transaction costs such as menu costs.

> Были разработаны теории "рациональной зарплаты" и "вязких жен", основанные на транзакционные изздержках (изджержках меню). 

This led to a New Synthesis model that again has a short-run/long-run dichotomy but
is now based on rigid microfoundations.

> Это привело к тому, что модели Нового Синтеща, которые как и раньше имели краткосрочную/долгосрочную дихотомию, но были основаны на жестких микрооснованиях. 

The Keynesian world is again a special case of
the general equilibrium

> Кейнсианский мир вновь стал особым случаем общего экономического равновесия.

PKs are highly critical of the rational-behaviour microfoundations and have pointed
out that the New Keynesian models have no independent role for investment expenditures and lack any meaningful role for finance. 

> Посткейнсианцы были крайне критичны об микрооснованиях рационального поведения экономических агентов и замечали, что новокейнсианские модели не имели независимой роли инвестиционных расходов и не имели значимой роли финансов.

Macroeconomic models with a central role
for income distribution have been refined by Post-Keynesians – building on the work of
Michał Kalecki.

> Макроэкономические модели с центральной ролью распределения дохода были уточнены посткейнсианцами на основе работ Michał Kalecki. 

The theories of endogenous money, financial instability and debt cycles
have been developed – building on the work of Nicholas Kaldor and Hyman Minsky.

> Теории эндогенных денег, финансовой нестабильности и долговых циклов были разработаны на основе работ Nicholas Kaldor и Hyman Minsky.

Today, the PKE paradigm is intellectually consolidated but academically marginalised.

> Сегодня паралигма PKE интеллектуально консолидирована, но академически маргинализовано. 

PKE is now a relatively coherent body of knowledge, with several textbooks, journals
and scholarly societies in several countries.

> PKE сейчас - относительно согласованное тело знаний, с множеством учебников, журналов и научных обществ во множестве стран.

However, the streamlining of the discipline
of economics, through journal ratings and, in the UK, the Research Excellence Framework, makes it ever more difficult to pursue an academic career as a Post-Keynesian or,
for that matter, as any kind of heterodox economist.

> Однако упорядочивание экономической дисциплины через журнальные рейтинги и (в Великобритании) через Research Excellence Framework делает сложнее строительство академической карьеры для посткейнсианцы, и если уж зашел об этом разговор, для любого рода экономиста-неортодокса. 

The financial crisis of 2008 has shaken the trust of the public in the economics profession, but not dented the self-confidence of economic orthodoxy. 

> Финансовый кризис 2008 встряс доверие публики в экономическую профессию, но не помял самоуверенности экономической ортодоксии. 

It has inspired the
interest of a new generation of critical students and, in many cases, also the broader
public in PKE.

> Он вдохновил интерес нового поколения критически настроенных студентов и во многих случаях широкой публики к PKE.

**Foundations: uncertainty, social conflict and institutions**

> **Основания: неопределенность, социальный конфликт и институты**

Modern mainstream economics is based on methodological individualism: the individual
is the basic unit of analysis.

> Современная мейнстримная экономика основана на методологическом индивидуализме: инидивид - простейшая единица анализа. 


Social and economic phenomena have to be explained from
individual choices and behaviour, with the standard assumption being that people are
rational and selfish (the so-called homo economicus). 

> Социальные и экономические феномены должны объясняться из индивидуального выбора и поведения со стандартными допушщения о том, что люди рациональные и эгоистичны (т.н. Homo Oeconomicus)

This microeconomic approach
has, since the 1970s, also become dominant in macroeconomics, where macro theories
are expected to be based on ‘microfoundations’.

> Микроэкономический подход, начиная с 1970-ых, стал доминирующим и в макроэкономике, где стало ожидаться, что макротеории должны покоится на строгих микрооснованиях. 

PKE, in contrast, rejects methodological individualism and takes a holistic and historic approach. It is based on sociology and
history rather than on the individual psychology of preferences or ‘utility’.

> PKE, в контраст с ним, отвергает методологический индивидуализм и берет на вооружение более цельный и исторический подход. Он скорее основан на социологии и истории, чем на индивидуальной психологии предпочтений или полезности. 

There are two main reasons why PKE rejects standard microfoundations.

>Сущесвтуют две причины, исходя из которых PKE отвергает стандартные микрооснования. 

First, PKE
highlights the concept of fundamental uncertainty, as contrasted to risk.

> Во-первых, PKE подчеркивает концепцию фундаментальной неопределенности, в отличие от риска.

Risk refers to
situations where the precise outcome is unknown, but the probability distribution is
known. 

> Риск отсылает к ситуациям, где точный исход неизвестен, но известно распределение вероятности. 

For example, in the case of rolling a dice, whilst we do not know in advance
what number will be thrown, we do know that each number has a one-sixth chance of
being thrown. 

>Например, в случае подбрасывания костей, в то время как нам неизвестно какой номер выпадет, но мы знаем, что вероятность выпадения каждого числа равна одной шестой. 

Fundamental uncertainty, in contrast, refers to situations where even such
probability distributions cannot be identified.

> Фундаментальная неопределнность в контраст отсылает к ситуациям, где распределение вероятности не может быть идентифицировано. 

To illustrate, think of an entrepreneur contemplating investment in a steel factory in 1936. 

> Чтобы проиллюстрировать, подумайте про бизнесмена, решаеющим стоил ли ему вкладываться в фабрику производства стали в 1936 году. 

The expected returns from that factory will depend on the demand for steel in
the coming years, which will depend, among other things, on whether there will be a
war, how long it will last and how economic performance will be after the war – not
to mention whether the factory itself will be destroyed by the war.

> Ожидаемая доходность от фабрики будет зависеть от спроса на сталь в будущие годы, который будет зависеть среди прочего от того, будет ли в будушие годы война или нет, как долго она будет длиться и насколько высок спрос будет после войны. Даже не упоминая будет ли сама фабрика разрушена в ходе войны. 

A rational entrepreneur would have to give expected values and probability distributions for all of these
scenarios – in a situation where World War II had not even begun!

> Рациональный бизнесмен должен был бы дать ожидаемые значения и вероятностные распределения для всех этих сценариев - в том числей той, гле Вторая мировая не произошла. 

The example may be
extreme, but PKE contends that investment decisions are usually made under conditions
of fundamental uncertainty.

> Это может быть и крайний сценарий, но PKE настаивает на позиции, что решения по поводу инвестиций сделаны в условиях фундаментальной неопределенности. 

Fundamental uncertainty is not a statement about the limited
cognitive ability of humans to make forecasts but a statement about the nature of social
reality.

> Фундаментальная неопределенность - это не предположение по поводу ограниченности когнитивносй способности человека делать предсказание, но предположение о природе социальной реальности. 

Societies do not work like a clock, where a future state can be readily deduced
from a present state, but rather involve human agency, emergent properties and irreversibilities.

> Общества не работают как час, где будущее состояние готово выводитя из нынешнего состояния, но включает в себя человеческое участие, выбор людей, эмерджентные характеристики и необратимости. 

The same is also true, therefore, of the economy.

> Это же также верно и для экономики.

So how do people make decisions in such an uncertain world? 

> Если так, то как же люди принимают решения в столь неопределенном мире?

First, Keynes, in a
famous phrase, uses the term ‘animal spirits’ to describe an innate drive to action rather
than inaction.

>Во-первых, Кейнс использует ставшую после знаменитую фразу, "животные духи", чтобы описать внутренний двигатель к действию, нежели к бездействию


In other words, he argues that there are non-rational, creative impulses
that are key to human behaviour.

> Другими словами, он утверждал, что существует нерациональные креативные импульсы, служащие ключом к человеческому поведению. 

Second, he highlights the importance of social conventions. 

> Во-вторых, он подчеркивает важность социальных коненций. 

People will look at what other people are doing and evaluate their behaviour in
light of these.

> Люди будут смотреть на то, что делают другие люди и оценивать свое поведение в свете поведения других людей.

Fundamental uncertainty is important for PK monetary theory because, as
we will see below, it gives rise to ‘liquidity preference’ and is also key for understanding
why investment expenditures tend to be highly unstable or, as Keynes says, appear to be
driven by ‘animal spirits’ rather than rational calculation.

> Фундаментальная неопределенность важна для посткейнсианской монетарной теории, потому как, что мы увидим ниже, она приводит к "предпочтению ликвидности". Это также ключ к пониманию, почему расходы на инвестирование склонны быть крайне нестабильными или, как говорил Кейнс, выглядят так будто они движимы скорее "животными духами", чем рациональной калькуляцией. 

The second PK objection to the individualist approach is that economics needs ‘macrofoundations’ at least as much as ‘microfoundations’.

> Вторая притензия посткейнсианства к индивидуалистическому подходу в том, что экономические "макрооснования" необходимы по крайней мере настолько же, насколько и "микрооснования".  

Societies are structured into social
groups, including classes, which are of particular importance to the economy.

> Общества структурируюися в социальные группы, в том числе классы, которые играет особенно важную роль в экономике. 

Classes are
social groups based on different functions within the economic process.

> Классы - это социальные группы, разделяемые по признаку, выполняемой ими функции внутри экономиеческого процесса. 

The most famous
ones are workers and capitalists, but the analysis can be extended to include rentiers and
shareholders or to give a class analysis of gender relations. 

> Самые знаменитые - это работники и капиталисты, но анализ может быть расширен, включив в себя рантье и акционеров или учесть классовый анализ гендерных отношений.

Members of a class have similar
interests, usually similar incomes and typically form their own political and economic
organisations to represent their interests.

> Члены класса имеют сходные интересы, обычно схожие доходы и типично формируют свои собственные политические и экономические организации, представляющие их интересы.

In capitalism, workers typically do not control
the factories and own little, if any, wealth. 

> При капитализме, рабочие типично не контролируют фабрики и владеюб малым, если вообще каким-то, богатством.

They have to find a job in a world where
involuntary unemployment is the norm, which puts them in a weak position relative to
capitalists. 

>Они вынуждены искать работу в мире, где вынужденная безработица - это норма, которая ставит из в более слабую позицию относительно капиталистов. 

Shareholders own firms or, more technically, the capital stock, and managers
control the production process and make the investment decisions that play a key role in
Keynesian macroeconomics.

> Акционеры владеют фирмами или, более формально, запасом капитала, и менеджеры контролируют процесс производства и совершают инвестиционные решения, играющие ключевую роль в кейнсианской экономике.

The conflict between capital and labour is a permanent feature of capitalist society,
and societies develop institutions to mediate and regulate the resulting class struggle.

> Конфликт между капиталом и трудом - это перманентная черта капиталистического общества и общества развивают институты, чтобы выступать в качестве посредника и регулировать результирующий классовый конфлик.

They also build welfare states and other social safety nets to cushion the extremes of
unequal income distribution, as part of social compromise. 

> Они также строят государства всеобщего благосостояния и другие социальные "сети безопасности", чтобы смягчать удары экстримальные исходов неравного распределения доходов, как часть общественного компромиса. 

In the PK view, social institutions are not mere distortions of a market process, but they play a constructive role in
mediating social conflict.

> С точки зрения посткейнсианства, общественные институты - не искажают рыночный процесс, а играют конструктивную роль в опосредовании социальных конфликтов. 

**The principle of effective demand**

> Принцип эффективного спроса


At the core of PK macroeconomics is the principle of effective demand: the idea that most
of the time our economies are demand-constrained (as opposed to supply-constrained)
so that the level of aggregate expenditures (on consumption, investment, government
spending and net exports) determines the level of output in an economy.

> В ядре посткейнсианской макроэкономики лежит принцип эффективног оспроса: идея того, что бОльшую часть времени наша экономика находится в рамках ограничения со стороны спроса (demand-constrained) (в контраст с ограничением со стороны предложения, или supply-constrained). 

Economic fluctuations, i.e. the business cycle, are driven by changes in expenditures, and in particular
investment expenditures. 

> Экономические фллуктуации, или бизнесс-цикл движим в изменениях в расходах, и в частноси инвестиционных расходах. 

This is in sharp contrast to the neoclassical approach, where the
idea of optimising agents always ensures that the economy is operating at full capacity,
such that supply constraints control how much the economy produces (that’s where the
phrase ‘supply-side’ comes from).

> Это резкий контарст с неоклассическим подходом, с его где идеей того, что оптимизирующие агенты обеспечивают то, что экономика всегда оперирует на полную мощность так, что ограничение предложения контролирует сколько экономика производит (это то откуда как раз и происходит понятие supply-side)

In both PKE and neoclassical theory, the goods market will be in equilibrium if investment equals saving: only at this point will income equal expenditures and the economy
can stably reproduce itself (ignoring for a moment state intervention and international
trade).

> И в PKE, и в неоклассической теории, рынок благ будет в равновесии, если инвестиции равны сбережениям: только в этой точке доход будет равен расходам и экономика будет стабильно себя воспроизводить (на секунду оставим за кадром государственные интервенции и международную торговлю)

According to neoclassical economics, the level of saving in the economy determines the level of investment, so that levels of investment are controlled by how much
people and businesses decide to save. 

> Согласно неоклассической экономике, урвоень сбережений в экономике определяет уровень инвестиций так, что уровень инвестиций контроллируется тем, как люди и фирмы решают сберегать. 

For PKE, the causality goes the other way around,
from investment to savings.

> Для PKE, логическая связь направлена в противоположную сторону: от инвестиций к сбережениям.

Investment creates new income via the multiplier process,
which in turn leads to additional saving. In this sense, investment is never constrained by
saving, but it can be constrained by the availability of finance and credit, as discussed below.

> Инвестиции создают новый доход через мультипликативный процесс, который в свою очередь приводит к дополнительным сбережениям. В этом смысле инвестиции никогда не ограничены сбережениями, но ограничены доступностью финансов и кредита. 

Investment is the most volatile component of gross domestic product (GDP), and it
is driven to a significant extent by animal spirits, i.e. by factors that cannot be reduced
to rational behaviour. 

> Инвестиции - эт осамый волатильный компонент внутреннего валового продукта (ВВП), и они в существенной мере движимы в "животными духами". Или другими словами, факторами, которые не могут быть сведены к рациональному поведению. 

Investment expenditures depend on the expected return on the
investments. The critical word here is ‘expected’. 

> Инвестиционные расходы зависят от ожидаемой доходности от инвестиций. Критическое словое здесь "ожидаемая".

The relevant rate of return is not a
technologically given ‘marginal product of capital’ (i.e. the additional quantity of goods
produced by adding an extra dollar’s worth of capital to the production process) but
is, rather, dependent on earnings expectations in the real world.

> Соответсвтующая норма доходности не дана нам технологически как предельный продукт капитала (или дополнительная единица производимого выпуска в результате добавления 1 $ дополнительного капитала в производственный процесс), но скорее зависит от ожиданй доходов в реальном мире. 

This is where fundamental uncertainty becomes important. 

> Это то, где фундаментальная неопределенность становится важной. 

The formation of expectations is not strictly a
rational process but a socio-psychological process that is prone to herd behaviour and
overreaction.

> Формирование ожиданий - не строго рациональный процесс, а процесс социо-психологический, который склонен к стадному поведению и непропорциональынм реакциям.

People tend to save a bigger portion of their income as they get richer. This phenomenon is important for PKE, because when people save more, less of the money gets
put directly into the economy as consumption.

> Люди склонны сберегать бОльшую долю своих доходов по мере того, как они становятся богаче. Этот феномент важен в рамках PKE, потому что когда люди сохраняют меньше, то меньше денег будут напрямую впрыснуты в экономику в качесвте потребления.

That means that income gains for poorer
people have a higher economic ‘multiplier’ effect than they do for richer people, as some
of the savings of richer people can sit idle and ‘leak’ out of circulation in the economy.

> Это значит, что прирост доходов бедных имеет более высокий экономический мултипликативный эффект, чем у богатых, т.к. некоторая часть себержений богатых может сидеть "без дела" или даже утекать за пределы циркуляции экономики.

Typically, when people receive additional money they will spend some of it and save
some of it – as opposed to spending or saving it all.

> Типично когда люди получают дополнительный доход, они потратят их часть, а некоторую сберегут. Мало кто будет тратить все на потребление или на сбережения. 

Keynes called this a ‘fundamental
psychological law’, and it is important because it means the multiplier for a national
economy will (almost) always be greater than 1 (meaning each extra dollar spent will
create more than 1 additional dollar of effective demand). 

> Кейнс называл это фундаментальным психологическим законом и он важен, так он значит, что мультипликатор для национальной экономики будет (почти) всегда выше единицы (что значит, что дополнительный доллар потраченный создает более чем 1 доллар эффективного спроса)

Therefore, in the PKE theory
an increase in investment expenditures will cause an increase in people’s incomes and thus saving, as part of the new income is saved. 

> Таким образом, в теории PKE возрастание в инвестиционных расходах вызовет рост доходов людей, а следовательно и сбережений, т.к. часть доходов сберегается. 

Consumption expenditure, then, plays a
passive role in Keynesian theory, and investment the active role, with national income
as the adjusting variable.

> Потребительские расходы затем играют пассивную роль в кейнсианской теории, а инвестиции - активную роль, где национальный доход - подстраивающаяся переменная. 

In practice, multipliers also depend on import propensities and on tax rates. 

> На практике, практике мультипликаторы также зависят от склонности к импорту и налоговых ставок. 

Modern
PKE typically has multipliers that allow for different consumption propensities out of
wage income and out of profit income because capitalists tend to be richer than workers.

> Современный PKE типично имел мультипликаторы, позволяющие различную склонности к потреблению исходя из типа дохода потребителя: склонность к потреблению с зарлпты, склонность к потреблению с прибыли, т.к. капиталисты богаче рабочих. 

The multiplier then depends on income distribution. 

> Мультипликатор затем зависит от распределения доходов.

In applied research, the openness of
an economy also has an important impact on the multiplier as trade openness represents
a ‘leakage’ in the multiplier process, as some of the new income leaves the national economic system

> В прикладных исследованиях, открытость экономики также исеет важное влияние, т.к. открытость к внешней торговле представляет собой "утечку" из экономики в рамках мультипликативного процесса, т.к. какие-то новые доходы покидают национальную эконмоическую систему. 

The goods market equilibrium is thus at the centre of Keynesian analysis. 

> Равновесие на рынке благ тем самым - центр кейнсианского анализа. 

Notice,
however, that this is a more limited notion of equilibrium than in neoclassical theory.

> Заметим, однако, что это более сдержанное утверждение равновесия, чем то, которое делает неоклассическая теория. 

This equilibrium is defined by incomes equalling expenditures, but unlike in neoclassical
economics, there is no claim that a market equilibrium is socially optimal. 

> Это равновесие определяется доходами приравненными к расходам, но в отличие от неоклассчиеской экономики, не делает утверждения о том, что рыночное равнвоесие общественное оптимаьно. 

Instead, within
the PKE conception, equilibrium in the goods market usually comes with involuntary
unemployment.

> Вместо этого, внутри коцепции PKE равновесие на рынке благ обычно имеет место с наличием вынужденной безработицы. 

**Unemployment and the futility of wage cuts**

> **Безработица и тщетность урезания зарплат**

The labour market is, in the PK analysis, a subordinate market. 

> Рынок труда - это в рамках посткейнсианского анализа подчиненный рынок.

Firms’ hiring decisions
will be determined most by their expected sales, with the cost of wages typically playing
a secondary role. 

> Решения фирм о найме будет определяться ожидаемыми продажами, в рамках которых затраты на зарплаты имеет второстепенную роль. 

The equilibrium level of output set on the goods market will thus also
determine the level of employment.

> Равновесный уровень выпуска устанавливаемый на рынке благ таким образом будет также определять и уровень занятости. 

This is also true in the short run for many mainstream models, and in particular for
the NAIRU (Non-Accelerating Inflation Rate of Unemployment) model, which theorizes that there is a hypothetical rate of unemployment at which inflation rates are stable.

> Это также верно и в краткосрочном периоде многих мейнстимных моделей, в частности для NAIRU-модели (Non-Accelerating Inflation Rate of Unemployment), которая теорезирует, что сущесвтует гипотетический уровень безработицы, при которой темп инфляции стабилен. 

However, mainstream NAIRU models are anchored in an equilibrium rate of unemployment, which is determined purely by supply-side considerations in the long run.

> Однако мейнстримные NAIRU-модели привязаны к равновесному уровню безработицы, которая определяется чисто стороной предложения в долгосрочном периоде. 

In
other words, they re-introduce the ‘classical’ rules of the game in the long run

> Другими словами, мы вновь добавляем "классические" правила игры в долгосрочном периоде. 

PKE, on the other hand, emphasises that labour markets are social institutions that
are adaptive even over longer time periods because of an endogenous labour supply or
social wage norms.

> PKE с другой стороны подчеркивает то, что рынки труда - это социальные институты, которые адаптивные в гораздо более долгих периодах по причине эндогенного предложения труда или социальной нормы зарплаты. 

This results in what is called path dependence or hysteresis: labour
markets have a ‘memory’, and increases in unemployment due to crisis can persist for a
long time.

> Это результирует в зависимости от предшествующего развития или хистерезиса: рынки труда имеют "память" и повышенная безработица вызванная кризисом может сохраняться долгие периоды времени.

In other words, PKE rejects the short-run/long-run dichotomy and argues
that employment is governed by demand even over long time horizons.

> Другими словами, PKE отвергает краткосрочную/долгосрочную дихотомию и утверждает, что занятость управляетя спросом даже в более долгие периоды времени. 

The NAIRU in
PKE is thus not an anchored ‘natural’ rate, but is endogenous and changes with respect
to demand shocks.

> NAURU таким образом в рамках PKE исходит не из естественной нормы безработицы, и меняется согласно шокам спроса. 

Debates about the long-run/short-run dichotomy aside, the difference between the
Keynesian and the neoclassical analysis can best be understood when considering the effects
of wage cuts in a recession.

> Отбросив дебаты о краткосрочной/долгосрочной дихотомии в сторону, разница между кейнсианском и неоклассическом анализами может быть лучше всего понята через пример о том, какой эффект урезания зарплат будут игать во время рецессии. 

PKs argue that wage cuts will usually be counterproductive.

> Посткейнсианцы утверждают обычно являются контрпродуктивными. 

The key question that Keynes raises is whether a wage cut will necessarily lead to an
increase in aggregate demand (in chapter 19 of the General Theory). 

> Ключевой вопрос, который поднимает Кейнс - это то, приведет ли снижение зарплат необходимо к росту аргегированного спроса (глава 19 Общей теории).

His answer to this is
‘no’. 

> Его ответ на него однозначный - нет. 

A wage cut will reduce the incomes of workers and, as a consequence, their consumption expenditure (which may also have a negative second-round effect on investment). 

> Урезание зарплат приведет к снижению дохоов работников и, как результат, их потребительских расходов (что также может иметь негативный эффект второго шага на инвестиции).

A wage cut could induce firms to invest more, if firms are very sensitive to wage
costs. 

> Урезание зарплат стимулирует фирмы инвестировать больше, если фирмы очень чувствительны к урезанию зарплат. 

However, that is unlikely in a recession because a recession means that firms already
have difficulties selling what they have produced, so why should they produce more just
because wages are falling?

> Этому маловероятно быть правдой во время рецессии, т.к. рецессия означает, что фирмы уже имеют сложности с продажами того, что они уже произвели, так что с какой стати им стоит производить больше только потому что они урезали зарплаты. 

It is more likely that firms will respond to the wage cut by reducing the prices they
are charging, but then what we get is deflation rather than an increase in production.

> Более вероятно, что фирмы отреагирует на урезание зарпат тем, что снизят цены товаров и услуг, но то, что в итоге мы будем иметь - это дефляция, а не рост выпуска. 

Deflation can have negative effects because it means that real debt burdens are increasing.

> Дефляция может иметь негативные эффекты, потому что означает, что реальная долговая нагрузка возрастет. 

With deflation, prices are falling, so the nominal income of firms will, for a given output
level, also be falling.

> С дефляцией падают цены так, что номинальные доходы фирм при данном уровне цен тоже будут падать. 

That is not necessarily a problem for firms in terms of financing their
running costs, because wages and, presumably, other input costs will also be declining.

> Это не необходимо проблема для фирм в смысле их текущих затрат, потому что зарплаты и, предположим, другие производственные расходы также будут снижаться.

But there is one key part of expenditures that is fixed in nominal terms: debt service.

> Но есть ключевая часть расходов, которая фиксирована в номинальных цифрах - обслуживание долга. 

During deflation, the value of debt and of interest payments on debt increases relative
to current income.

> Во время дефляции, стоимость долга и оплата процентов по долгу возрастает относительно текущего дохода. 

Thus the more heavily indebted firms and households are, the more
dangerous are the effects of deflation.1

> Таким образом чем более налогове бремя фирм и домохозяйств возрастает, тем более опасны эффекты дефляции. 

Keynes discusses no fewer than eight channels by which a wage cut may affect
demand (and thus employment) and identifies just one that can lead to positive effects
during a recession: a wage cut will increase net exports if trade partners do not experience a similar wage cut.

> Кейнс привел не менее, чем 8 каналов, по которым урезание зарплат может оказать влияние на спрос (и тем самым занятость) и идентифицирует только, который может привести к позитивным эффектам во время рецессии: урезание зарплаты приведет к поднятию чистого экспорта, если торговые партнеры не переживают схожие урезания зарплат. 

In practice the effectiveness of this channel will depend
on how open the economy is to trade and what the economic condition is in other
countries.

> На практике эффективность этого канала зависет от того, насколько открыта экономика для внешней торговли и в каком экономическом положении находятся другие страны.

It can work in a small open economy, when the rest of the world is growing
healthily. Overall the verdict, however, is clear: a wage cut in a recession will normally not reduce unemployment. 

> Это может сработать для небольшой открытой экономики, когда остальной мир находится в фазе значительного экономического роста. Но в целом вердикт ясный: урезание зарплат в нормальных условиях не приведет к снижению безработицы. 

What is often euphemistically called ‘labour market
reform’ will not work.

> То что часто называют эвфемизмом "реформа рынка труда" не работает!

Modern PKE thus distinguishes between wage-led and profit-led demand regimes.

> Современная PKE делает различие между age-led и profit-le режимами спроса. 

In these models, an increase in wages will positively affect consumption but can have a
negative impact on investment and (assuming that wage costs are stable in the trading
partners) net exports. 

> В таких моделях, рост зарплат оказывает положительное влияние на потребление, но может иметь негативное влияние на инвестиции и (предполагаем, что расходы на зарплату стабильны среди торговых партнеров) чистый экспорт. 

The relative size of these effects then determines whether the economy is overall wage-led or profit-led. 

> Относительный размер этих эффектов затем определяет будет ли экономика в целом wage-led или profit-led. 

In a wage-led demand regime, the consumption
effect overpowers the investment and net export effect. 

> В wage-led режиме спроса эффект потребления перевешивает эффект инвестиций и чистого экспорта. 

This means an increase in wages
will increase total demand and employment. 

> Это значит, что возрастание в зарплатах приводит к возрастанию агрегированного спроса и занятости. 

In a profit-led regime the opposite is true –
wage increases dampen investment and net exports enough to have a negative effect on
demand and employment.

> В profit-led режиме спроса противоположное верно - рост зарплат приведет к снижению инвестиций и чистого экспорта, что приведет к в целом негативному эффекту на спрос и занятость. 

There is an important distinction between the domestic effect
and open economy effects.

> Существует важное различие между национальным эффектом и эффектами открытой экономики. 

While individual countries can export themselves out of a
crisis via exports, the world economy overall can’t do that, because it is a closed economy
(i.e. we can’t rely on increased exports to Mars).

> Пока отдельные страны могут "экспортировать себя из кризиса", мировая экономика в целом сделать этого не может, потому что это закрытая экономика (мы пока не можем полагаться на возрастание экпорта на Марс)

Most PKE expect actual economies
to be wage-led, at least domestically, and there is presently a lot of research to identify
demand regimes empirically

> Большиснтво посткейнсианцев ожидают, что реально сущесвтующие экономики wage-led по крайней мере наионально, и в наши сущесвтует множество исследований, которе пытаются индентифицировать режим спроса эмпиричеки. 

**Monetary theory: endogenous money, liquidity
preference and financial cycles**

> **Монетарная теория: эндогенные деньги, предпочтения ликвидности и финансовые циклы**

Keynesian theory of fundamental uncertainty has important implications not only for
the theory of investment but also for monetary theory. 

> Кейнсианская теория фундаментальной неопределенности имеет важные следствия не только для теории инвестиций, но также и для монетарной теории. 

While in mainstream/neoclassical economics, agents hold money essentially to buy goods and services (often called
‘transactions demand’ for money), in PKE money is held as a financial asset, rather than
invested for potential returns. 

> В то время как в рамках мейнстримной и неоклассической экономики, агенты хранят деньги в сущности для того, чтобы покупать товары и услуги (т.н. транзакционный спрос на деньги), в PKE деньги зранятя как финансовые активы, нежели инвестированы для потенциальной доходности. 

This may at first sound counterintuitive: why would anyone hold money which gives no yield? 

> Это на первый взгляд может звучать контринтуитивно: почему кто-либо будет держать деньги, не собирая с них процента?

The answer is that in a world with uncertainty,
what money offers is liquidity, i.e. flexibility. Holding money means earning no (or very
little) interest, but it ensures the ability to buy assets in the next period.

> Ответ в том, что в мире с неопределенностью, то что предлагаю деньги - это ликвидность. Иными словами гибкость. Хранить деньги значит не получать никакого (или почти никакого) процента, но это дает возможность купить активы в следующем периоде. 

Other financial
assets will offer higher returns, but they come with a risk of capital losses (or positively
put: capital gains) and with the risk of the asset becoming illiquid.

> Другие финансовые активы часто предлагают более высокую доходность, но они несут в себе риск потери капитала и с риском того, что актив станет неликвидным. 

In normal times, this
will mean that people will want to minimise their cash holding, because markets are
liquid and reasonably predictable.

> В нормальные времена, это будет значить, что люди пытаются минимизировать наличность "на руках", потому что рынки ликвидны и разумно предсказуемы. 

However, in times of financial distress the demand for
money, what Keynes called the ‘liquidity preference’, will go up sharply.

> Однако, во времена финансовых неурядиц спрос на деньги то, что Кейнс называет "предпочтение ликвидности", резко возрастет. 

Investors will
hold money as an insurance against an uncertain future.

> Инвесторы будут держать деньги как страховку от неопределенного будущего. 

In a financial crisis there will be
a rush to liquidity – there will be a run on banks. 

> В финансовые кризисы будет происходить гонка за ликвидностью - произойдет налет на банки. 

This is also known as a liquidity trap:
a situation when the public wants to hold money, independent of the interest rate. 

> Это также известно как ловушка ликвидности: ситуация в которой публика хочет держать деньги в независимости от ставок процента. 

As a
consequence other financial markets tend to freeze, such as happened in 2008 when
interbank lending and commercial short-term lending and various other markets ceased
functioning.

> Как следствие, другие финансовые рынки склонны к тому, чтобы быть замороженными, как то случилось в 2008 когда межбанковский кредит и коммерческое краткосрочное кредитование и другие различне рынки прекратили свое функционирование. 

But where does the money come from in the first place? 

> Но откуда беруться деньги изначально?

Money in a modern economy is essentially bank deposits. 

> Банки в соврменной экономики - это в сущности банковские депозиты. 

In other words, money is the short-term liabilities of the
banking sector. 

> Другими словами, деньги - это краткосрочные пассивы банковской системы.

It’s an asset for everyone else (the non-financial sector) but a liability for
the banks. 

> Это активы для всех остальных (нефинансового сектора), но пассив для банков.

The banking sector creates money as a side effect of its lending decisions. 

> Банковский сектор создает деньги как побочный эффект его решений по кредитованию. 

Typically, when banks extend credit, somewhere in the system deposits, i.e. money, will be
created. 

> Типично, когда банки расширяют кредитование, где-то в системе создаются депозиты, т.е. будут созданы деньги.

This notion of money also implies that if loans are re-paid, money is destroyed.

> Из этого также следует то, что если кредит выплачен, то деньги уничтожаются.

The lending decisions of banks will depend on their assessment of the creditworthiness
of the applicants, which in turn will depend on their expected income streams and the
market valuation of their collateral.

> Решения банков о кредитовании будет зависеть от их от оценкок кредитоспособности подающих заявки, что в свою очередь будет зависеть от ожидаемого потока дохода и рыночной стоимоисти их залога. 

Central banks in this view do not control the money supply, as is taught in most
textbooks, but rather set the base interest rate.

> Центральные банки на их взгляд не контролируют предложение денег, как то учат большинство учебников, но скорее задают ставку процента. 

They lend to banks and thereby create
an anchor for the hierarchy of interest rates that the banks charge. 

> Они занимают банка и таким образом стоят у руля ставки процент, которые берут банки. 

The surcharge (above
the central bank’s base rate) that banks will charge for interest on loans will depend on
the riskiness of loans and perceived uncertainty.

> Надбавка (выше ставок центарльного банка), которые берут банки будет зависеть от рискованности кредитов и воспринимаемой неопределенности. 

Central banks can also use their balance
sheets to buy financial assets, as has happened as part of Quantitative Easing, but this does
not necessarily translate into an increase in circulating money (as measured e.g. by M2)
if banks do not increase their lending.

> Центральные банки такде могут использовать свои балансы, чтобы покупать финансовые активы, что происходит, например, во время количесвтенных смягчений, но это совсем необязательно трансформируется в возрастание циркулирующих денег (измеряемых в М2), если банки не увеличивают свою кредитование. 

The PK view of endogenous money contrasts
sharply with the Monetarist theory of an exogenous money supply under the control of the central bank.

> Точка зрения посткейнсианцев на деньги как эндогенную переменную жетско контрастирует с взглядом монетаристской теории, в рамках которой предложение денег полагается экзогенным и под контролем центрального банка. 

Monetarists believe that the central bank can control so-called base
money and that the money multiplier (the relationship between base money and deposits) is stable. 

> Монетаристы верят, что центральный банк может контролировать так называемую монетарную базу и что денежный мультипликатор (отношение между денежной базой и депозитами) стабильна. 

This model is still widely used in macroeconomics textbooks, but it is
increasingly out of line with how central banks themselves conceive of the process of
money creation

> Эта модель все еще широко используется в учебниках по макроэкономики, но она возрастающе расходятся с тем как сами центральные банки видят себя в рамках процесса создания денег. 

The theory of endogenous money highlights the flexibility of the financial system.

> Теория эндогенных денег подчеркивает гибкость финансовой системы. 

If a creditworthy firm has a reasonable investment project, the financial sector can, and
in normal times will, provide the finance for investment. 

> Если кредитоспособные фирмы имеют разумные инвестиционные проекты, то финансовый сектор может в нормальное время предоставлять финансы для инвестиций. 

It is in this sense that investment is never constrained by savings but can be constrained by finance.

> Это понимается в том смысле, что инвестиции никогда не ограничены сбережениями, но могут быть ограничены финансами. 

If banks are
worried about their balance sheet or don’t share the expectations of the firm, credit may
be restricted and firms’ investment may indeed be constrained.

> Если банки беспокоятся о своих баланасах или не разделяют ожиданий фирмы, кредит может быть ограничен и инвестиции фирмы действительно будут ограничены.

The flexibility of the
financial sector, however, is a double-edged sword as it will often lead to pro-cyclical
movements in credit that amplify the business cycle, when banks share the same broad
economic outlook as firms, i.e. when the animal spirits of firms and banks are aligned.

> Гибкость финансового сектора однако - обоюдоострый клинок, так как он часто приводит к процикличесским движениях в кредитовании, который усиляет бизнес-цикл, когда банки разделяют широкий экономический взгляд фирм, или другими словами когда животные духи фирм и банков совпадают. 

In
practice, the market value of collateral plays a key role in the assessment of creditworthiness. 

> На практике, рыночная оценка залога играет ключевую роль в оценке кредитоспособности.

This also means that financial bubbles can easily inflate, as credit growth and asset
prices (or real estate prices) form a positive feedback loop.

> Это также значит, что финансовые пузыри могут просто раздуться пока рост кредитов и цен активов (или цены недвижимости) формируют положительную обратную связь.

Higher property prices lead to
more collateral and higher lending; rapid credit growth fuels the property prices bubble.

> Повышенные цены на собственность приводят к росту залогов и повышению кредитования; быстрый рост кредита питает ценовой пузырь недвижимости. 

Hyman Minsky was one of the leading thinkers to develop Keynes’ financial analysis into
a theory of debt cycles and financial instability.

> Хайман Мински был ондим из лидирующих мыслителей, который развили и расширили финансовый анализ Кейнса в теорию циклов долга и финансовой нестабильности.  

**Economic policy**

> **Экономическая политика**

In the PK vision, capitalist market economies are prone to large fluctuations. 

> С точки зрения посткейнсианства капиталистические рыночные экономики склонны к масштабным флуктуациям. 

Situations
of full employment, while possible, will only occur by chance.

> Ситуация полной занятости возможно, но возникают случайно. 

PKE thus advocates
extensive government involvement, but PKE is not associated with a particular political
philosophy.

> PKE таким образом поддерживает большое государственное вмешательство, но PKE не ассоциируется с определенной политической философией. 

While Keynes was a liberal, Kalecki was a socialist, who returned from the
West to work in communist Poland in the 1950s. 

> В то время как Кейнс был либералом, Калецки был социалистом, вернувшимся с Запада на работу в коммунистическую Польщу в 1950-ых годах. 

What PKs share is the sentiment that
investment decisions should not be left to the market. 

> Что разделяют все посткейнсианцы - это сентимент о том, что инвестиционные решения не должны быть целиком отданы на усмеотрение рынка. 

Keynes famously, but opaquely,
referred to this as ‘the socialisation of investment’. 

> Кейнс как известно непрозрачно отсылал к этому как к "социализации инвестиций". 

The more liberally inclined would aim
to achieve this by monetary and fiscal policy, but it may also be attempted by industrial
policy, credit allocation or via state-owned enterprises. 

> Более либерально настроенные посткейнсианцы целят на остижение этого путем монетарной и фискальной политики, а также промышленной политикой и аллокацией кредита через государственные предприятия. 

In the following we outline policy proposals that follow from PK analysis and contrast them to the current mainstream
view on economic policy. 

> Далее мы наместим очертания политических рекомендаций, которые следуют из посткейнсианского анализа, а также наметим их контраст с нынешним мейнстримным взглядом на экономическую политику. 

This is for illustration, but it should be kept in mind that PKE
is a set of theories about the working of capitalist economies, not a political ideology.

> Это иллюстрация, но нам следует держать в уме, что  PKE - это мнодество теорий р работе капиталистической экономики, а не политическая идеалогия. 

As mentioned above, in PKE market economies will not, in general, lead to full
employment.

> Как ранее упомяналось, с точки зрения PKE рыночные экономики не будут как правило двигаться в сторону полной занятости. 

Employment is determined by effective demand, which may fluctuate
depending on changes in animal spirits and financial crises.

> Занятость определяется эффективным спросом, который может испытывать флуктуации, заисящие от изменения в животных духах и финансовых кризисов. 

This is the basic PK argument
for extensive government intervention.

> Это простейший посткейнсианский аргумент для сущесвтенных государственных интервенций. 

Governments should use fiscal and monetary
policy to increase demand and approximate full employment. 

> Правительствам стоит использовать фискальную и монетарную политику для повышения спроса и приближения к полной занятости. 

Fiscal policy plays the key role in this approach because in times of crises, in particular after financial and debt overhang crises, monetary policy will be limited in its effectiveness.

> Фискальная политикак играет ключевую роль в данном подходе, потому что во времена кризиосов, в частности после финасновых и долговых кризиосв, монетраная политика будет ограничена в своей эффективности. 

From this follows a view of government budget deficits that differs sharply from the
mainstream. 

> Из этого следует взгляд на бюджетные дефициты жестко отличный от мейнстримной экономики.  

In an economic crisis, firms reduce their investment sharply and households
will try to increase savings.

> Во время экономических кризисов, фирмы жескто сокращают свои инвестиции и домохозяйсвта будут пытаться повысить свои сбережения. 

As a result the private sector will want to save more and
invest less – it moves to a net lending position.

> Как результат, частный сектор будет желать сберегать больше и инвестировать меньше. Он двигает в сторону нетто-кредитора. 

There are two ways this can be resolved:
(1) either national income collapses such that private savings can equal investment, with
mass unemployment resulting; or (2) some other sector has to move into a net borrowing
position (and thereby stabilise national income). 

> Это может быть решено двумя способами:
(1) или национальные доходы рухнут настолько, что частные сбережения приравняются к инвестиция с массовой безработицей как результатом 
или (2) некоторые сектора должны сдвинуться в позицию нетто-заемщика (и таким образом стабилизировать доход)

This is why government budget deficits
in a crisis are a desirable outcome: they substitute the lacking private expenditures and
thereby prevent unemployment. 

> Это то, почему бюджетые дефициты государства во времена кризисов - это желаемый исход: они замещают недостающие частные расходы и таким образом предотвратить безработицу. 

How long should the government run deficits?

> Как долго стоит правительствам держать дефициты? 

 The
short PK answer is: as long as there is substantial unemployment.

> Короткий ответ посткейнсианцев: так долго, сколько существует значительная безработица. 

PKE regards financial markets as a potential source of instability because endogenous
money creation can lead to boom-bust cycles and sustain asset price bubbles.

> PKE считают, что финансовые рынки являются потенциальным источником нестабильности. потому что эндогенное создание денег могут приводить к бум-баст циклам и поддерживать пузыри активов. 

Central
banks thus have to use countercyclical monetary policy. 

> Центральные банки таким образом должны использовать контрциклическую монетарную политику. 

The main instrument for this
would be increased macroprudential regulation rather than interest rate policy.

> Главный инструмент для этого - скорее макропруденциальные регуляции, чем политика ставки процента.  

Most PKs
are also sceptical about the benefits of international capital flows.

> Большинство посткейнсианцев также скептичны по поводу преисуществ международного движения капитала. 

International capital
flows often reinforce financial bubbles (e.g. the Irish and Spanish real estate bubbles were
importantly fuelled by capital inflows).

> Международные движения капитала подкрепляют финансовые пузыри (как напрммиер ирландские и испанские пузыри недвижимости, котоыре были в том числе напитаны иностранными потоками капитала).

Thus capital management or capital controls are
part of the PK repertoire.

> Таким образом управления капиталом или контроль над капитаом являются частью репертуара посткейнсианцев.

Control of domestic monetary conditions is a precondition for
effective national policies.

> Контроль над внутренним мнетранымы условиями - это предпосылка эффективной национальной политики. 

From the PK view there is no reason why monetary policy should be primarily concerned with inflation.

> С точки зрения посткейнсианцев не сущесвтует никакой причины, почему монетарная политика должны быть только сфокусирована на инфляции.  

Rather, central banks should be concerned with financial stability
and with the overall level of economic activity, in particular the level of unemployment.

> Скорее центральным банкам следует быть озабоченными финансовой стабильностью и с уровнем экономической активности по экономике в целом, в частности уровнем безработицы. 

So how would PK tackle inflation?

> Как же кейнсианцы будут справляться с инфляцией?

For PKs (leaving aside the issue of inflation caused
by price changes of imported goods like oil) the source of inflation is, most of the time,
unresolved distributional conflicts.

> Для посткейнсианцев (оставляя в стороне проблему инфляции, вызванная измененим цен импортированных благ таких, как нефть) источник инфляции - это в большиснтве случаев конфликт в распределении. 

Thus the government should try to support social
settlements that entail distributional compromises between capital and labour. 

> Таким образом правительству следует пытаться поддердивать социальные расходы, чтобы поддерживать распределительный компромисс между трудом и капиталом. 

Wagebargaining institutions thus become useful social institutions, where social compromises
can be forged, rather than a mere ‘market imperfection’.

> Институции, проводящие переговоры по оплате труда, таким образом становятся полезными социальными институтами, где может быть достигнут социальный компромисс, а не рыночное несовершенство. 

PK theory also has important implications for international economic relations, particularly for international monetary arrangements like the Euro area. 

> Посткейнсианская теория такде иимеет важные следствия для международных экономических отношений, в частности интернациональные монетраные соглашения такие как например зона Евро. 

In these arrangements,
when there are countries running trade surpluses and countries running trade deficits,
the burden of adjustment falls entirely on the deficit countries. 

> В рамках таких соглашений страны могут иметь как торговые дефициты, так и торговые профициты, бремя подстройки полностью ложиться на страны с торговым дефицитом. 

Such a system, which is
actively enforced by the European Commission and the infamous Troika, results in a
deflationary bias – i.e. those countries with current account deficits will have to reduce
demand and/or suppress wages in order to gain competitiveness.

> Такая система, которая активна проводится Европейской Комиссией и печально известной Тройкой, результирует в дефляционном предубеждении -  т.е. эти страны с текущими торговым дефицитом будут снижать спрос и/или снижать зарплаты с целью получить больше конкурентоспособности. 

If, on the other hand,
those countries with current account surpluses (in the European case, Germany) were to
adjust instead, they would have to stimulate domestic demand and foster wage growth
to reduce their trade surpluses. 

> Если, с другой стороны, страны с текущими профицитом торгового баланса (в случае ЕС, Германия) должны были бы подстраиваться, то они стимулировали внутренний спрос, чтобы стимулировать внтуреннйи рост зарплат для снижения их торгового дефицита. 

Keynesian international arrangements would be such as
to foster a balanced adjustment where surplus countries carry a substantial part of the
burden of adjustment.

> Кейнсианские международные соглашения таким образом должны быть такими, чтобы перекладывать грух подстройки на страны с торговым профицитом. 

**Note**

1.  The issue of deflation receives a somewhat schizophrenic treatment in modern mainstream economics. Standard macro textbooks typically assume an exogenous money supply. Deflation, i.e. a fall in prices, will then increase the real money supply and have an expansionary effect. Most modern central banks, however, are quite concerned about the danger of deflation and think that it has negative demand effects.

> 1. Проблема дефляции получает довольно шизофреничное отношений в современной сейнстримной экономике. Стандартные учебники по макроэкономики типично предпологают предложение денег экзогенным. Дефляции (падение цен) таким образом приведет к повышению реального предложению денег и будет иметь экспансионный эффект. Большинство центральных банков однако обеспокоены опасностями дефляции и видят ее негативные эффекты на спрос.
