## Введение 

В Python есть довольно много GUI фреймворков (graphical user interface), однако только Tkinter встроен в стандартную библиотеку языка. У Tkinter есть несколько преимуществ. Он кроссплатформенный, поэтому один и тот же код можно использовать на Windows, macOS и Linux. 

Хотя Tkinter является популярным GUI фреймворком на Python, у него есть свои недостатки. Один из них заключается в том, что графические интерфейсы, созданные с использованием Tkinter, выглядят устаревшими. Если вам нужен современный, броский интерфейс, то Tkinter может оказаться не совсем тем, для этого есть PyQt5 который развивается сильнее в данном плане.

Tk - это библиотека виджетов, имеющая корни в языке программирования Tool Command Language (Tcl). Оба эти инструмента были разработы John Ousterhout в 1980-ых для упрощения программирования для инженеров. Как и сам Python Tcl/Tk возник на Unix-системе и уже позже мигрировал на другие ОС такие, как Windows и macOS. 

**Tkinter** - это Python интерфйс *Tk* GUI библиотеки. 

**Преимущества Tkinter:**
* Это стандартная библиотека 
* Он стабилен - у него медленный девеломпмент и эволюционный
* Это всего лишь GUI toolkit - он не имеет threading library, network stack или filesystem API
* Он простой и понятный. 

**Негативные аспекты Tkinter:**
* Он выглядит и ощущается устаревшим.
* Не способен на более комплексные виджеты. Не имеет:rich text editors, 3D graphics embedding, HTML viewers, или specialized input widgets

* **Родительский виджет** - это виджет, на котором ... Виджеты в tkinter ранжированы иерархично, каждый виджет содержиться в другом вплоть до корневого окна

Данный метод также является т.н. **geometry manager method** Его задача в том, чтобы определить то, как виджет будет присоединен к родителському виджету и нарисовать его. Без него виджет будет существовать, но его не будет видно на окне


* ``root.mainloop()`` - как правило последняя строка практически любой приложения tkinter. Она начинает **event loop** приложения.
* **event loop** - это бесконечный цикл, который непрерывно процессит любой *ivent* (событие), который происходит во время execution этого приложения.
* События: движения мышкой, нажатия или другие сгенерированные пользователем виды активности. 
* Этот цикл продолжается до тех пор, пока программа не выйдет, так что любой код после этой строки не будет запущен, пока главное окно программы не закрыто. По этой причине данная строка обычно последняя.

````python
data = sheet.values

# Indicate the columns in the sheet values
cols = next(data)[1:]

# Convert your data to a list
data = list(data)

# Read in the data at index 0 for the indices
idx = [r[0] for r in data]

# Slice the data at index 1 
data = (islice(r, 1, None) for r in data)

# Make your DataFrame
df = pd.DataFrame(data, index=idx, columns=cols)
````
